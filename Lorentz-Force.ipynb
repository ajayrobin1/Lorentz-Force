```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
```

*Both electric and magnetic fields are uniform and constant.

*Magnetic field is along the z-direction.

```python
x0, y0, z0=0, 0, 0 # initial position

Bz = 0 #Tesla

Ex, Ey, Ez=0, 0, 1 # N/m

#initial velocity
vx0, vy0, vz0 = 1,0,0 # m/s

def dsdt(t,s):
    x, y, z, vx, vy, vz = s
    return [vx,
            vy,
            vz,
            -Ex + (2 * vy * Bz),
            -Ey - (2* Bz * vx),
            -Ez
           ]

# the charged particle is moving along the y-direction with uniform velocity

s0 = (x0, y0, z0, vx0, vy0, vz0)

t = np.linspace(0, 100, 1000)
sol = odeint(dsdt, y0=s0, t=t, tfirst=True)

xpos = sol.T[0]
ypos = sol.T[1]
zpos = sol.T[2]
```

```python
fig = plt.figure(figsize = (6,6))

# v1 = np.array([1, 2, 3])
# v2 = np.array([-2, 1, 4])


ax = plt.axes(projection ='3d')

ax.plot3D(xpos, ypos, zpos, 'green')

ax.set_title('Lorentz Force: B=0 , E along z-axis and velocity along x-axis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax.grid(False)

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
     
plt.show()
```
