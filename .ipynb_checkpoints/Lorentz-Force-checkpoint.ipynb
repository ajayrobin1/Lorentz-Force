```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

*Both electric and magnetic fields are uniform and constant.

*Magnetic field is along the z-direction.

```python
x0, y0, z0=0, 0, 0 # initial position

Bz = 0.1 #Tesla

Ex, Ey, Ez=0, 0, 0 # N/m

def dsdt(t,s):
    x, y, z, vx, vy, vz = s
    return [vx,
            vy,
            vz,
            Ex + (2 * vy * Bz),
            Ey - (2* Bz * vx),
            Ez
           ]
#initial velocity
vx0, vy0, vz0 = 0, 1, 1 # m/s

s0 = (x0, y0, z0, vx0, vy0, vz0)

t = np.linspace(0, 100, 1000)
sol = odeint(dsdt, y0=s0, t=t, tfirst=True)

xpos = sol.T[0]
ypos = sol.T[1]
zpos = sol.T[2]
```

```python
fig = plt.figure(figsize = (6,6))

ax = plt.axes(projection ='3d')

ax.plot3D(zpos, xpos, ypos, 'green')

ax.set_title('Charged Particle in E-B field')
ax.set_xlabel('Z axis')
ax.set_ylabel('X axis')
ax.set_zlabel('Y axis')

plt.show()
```
