```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.special import jv
import matplotlib as mpl
from matplotlib.ticker import LinearLocator
```

```python
D = 1 # *(distance of the screen) m
k = 2 * np.pi * 10**6 # (wave number) /m
a = 0.01 # m (radius of the apperture)
```

The intensity of the incident light is 1 unit.

```python
def get_intensity(x, D):
    theta_0 = np.arctan(x/D)
    j1 = jv(1,k * a* np.sin(theta_0))
    intensity = (2* j1 / (k* a * np.sin(theta_0)))**2
    return(intensity)
r = np.linspace(0.00001,0.0001,10000)
ir1 = get_intensity(r, 0.1)

plt.plot(r,ir1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(0.0001, 1, 1000)
p = np.linspace(0, 2*np.pi, 100)
R, P = np.meshgrid(r, p)
Z = get_intensity(R, 1)

# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)

# Plot the surface.
ax.plot_surface(X, Y, Z, cmap=plt.cm.YlGnBu_r)

# Tweak the limits and add latex math labels.
ax.set_zlim(0, 1)
ax.set_xlabel(r'$\phi_\mathrm{real}$')
ax.set_ylabel(r'$\phi_\mathrm{im}$')
ax.set_zlabel(r'$V(\phi)$')

plt.show()
```
